# **Task 4. Feature Matching**

```python
import cv2
import matplotlib.pyplot as plt

image1 = cv2.imread("eminem.jpg", 0)
image2 = cv2.imread("eminem2.jpg", 0)

height1, width1 = image1.shape[:2]
height2, width2 = image2.shape[:2]

if height1 != height2 or width1 != width2:
    image2 = cv2.resize(image2, (width1, height1))
```

```python
sift = cv2.SIFT_create()

keypoints1, descriptors1 = sift.detectAndCompute(image1, None)
keypoints2, descriptors2 = sift.detectAndCompute(image2, None)

bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck = True)

matches = bf.match(descriptors1, descriptors2)

matches = sorted(matches, key = lambda x: x.distance)

image_matches = cv2.drawMatches(image1, keypoints1, image2, keypoints2, matches[0:10], None, flags = cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(image_matches, cv2.COLOR_BGR2RGB))
plt.title("Feature Matching with SIFT")
plt.axis("off")
plt.show()
```
