# **Machine Problem 3: Feature Extraction and Object Detection**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("building.jpg")
image1 = cv2.imread("building1.jpg")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

# SIFT: Extract Keypoints and Descriptors

```python
sift = cv2.SIFT_create()

keypoints, descriptors = sift.detectAndCompute(gray, None)
keypoints1, descriptors1 = sift.detectAndCompute(gray1, None)
```

```python
sift_result = cv2.drawKeypoints(image, keypoints, None)
sift_result1 = cv2.drawKeypoints(image1, keypoints1, None)

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(sift_result, cv2.COLOR_BGR2RGB))
plt.title("SIFT Keypoints")

plt.subplot(1, 2, 2)
plt.title("SIFT Keypoints")
plt.imshow(cv2.cvtColor(sift_result1, cv2.COLOR_BGR2RGB))
plt.show()
```

# ORB: Extract Keypoints and Descriptors

# SIFT and BF: Feature Matching with Brute-Force and FLANN

```python
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck = True)

matches = bf.match(descriptors, descriptors1)

matches = sorted(matches, key = lambda x: x.distance)

sift_bf_matches = cv2.drawMatches(image, keypoints, image1, keypoints1, matches[0:10], None, flags = cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

# SURF and BF: Feature Matching with Brute-Force and FLANN

```python
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck = True)

matches1 = bf.match(descriptors2, descriptors3)

matches1 = sorted(matches1, key = lambda x: x.distance)

surf_bf_matches = cv2.drawMatches(image, keypoints2, image1, keypoints3, matches[0:10], None, flags = cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

# ORB and BF: Feature Matching with Brute-Force and FLANN

```python
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck = True)

matches2 = bf.match(descriptors4, descriptors5)

matches2 = sorted(matches2, key = lambda x: x.distance)

orb_bf_matches = cv2.drawMatches(image, keypoints4, image1, keypoints5, matches[0:10], None, flags = cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

# SIFT and FLANN: Feature Matching with Brute-Force and FLANN

```python
FLANN_INDEX_KDTREE = 1
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
search_params = dict(checks=50)

flann = cv2.FlannBasedMatcher(index_params, search_params)

# Match descriptors
matches = flann.knnMatch(descriptors, descriptors1, k=2)

good_matches = []
for m, n in matches:
    if m.distance < 0.7 * n.distance:
        good_matches.append(m)

sift_flann_matched = cv2.drawMatches(image, keypoints, image1, keypoints1, good_matches, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

ORB and FLANN: Feature Matching with Brute-Force and FLANN

```python
# Ensure descriptors are of type float32
descriptors4 = descriptors4.astype(np.float32)
descriptors5 = descriptors5.astype(np.float32)

FLANN_INDEX_KDTREE = 1
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
search_params = dict(checks=50)

flann = cv2.FlannBasedMatcher(index_params, search_params)

# Match descriptors
matches3 = flann.knnMatch(descriptors4, descriptors5, k=2)

good_matches = []
for m, n in matches3:
    if m.distance < 0.7 * n.distance:
        good_matches.append(m)

orb_flann_matched = cv2.drawMatches(image, keypoints4, image1, keypoints5, good_matches, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

# FLANN Feature Matching OUTPUT

```python
plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 1)
plt.imshow(cv2.cvtColor(sift_flann_matched, cv2.COLOR_BGR2RGB))
plt.title("FLANN Feature Matching with SIFT")
plt.axis("off")

plt.subplot(2, 1, 2)
plt.imshow(cv2.cvtColor(orb_flann_matched, cv2.COLOR_BGR2RGB))
plt.title("FLANN Feature Matching with ORB")
plt.axis("off")
plt.show()
```

# Image Alignment Using Homography

```python
bf = cv2.BFMatcher(cv2.NORM_L2)
matches = bf.knnMatch(descriptors, descriptors1, k = 2)

good_matches = []
for m, n in matches:
    if m.distance < 0.8 * n.distance:
        good_matches.append(m)

src_pts = np.float32([keypoints[m.queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
dst_pts = np.float32([keypoints1[m.trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)

M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 2.0)

h, w, _ = image.shape
result = cv2.warpPerspective(image, M, (w, h))
```

```python
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Image Alignment using Homography")
plt.axis("off")
plt.show()
```
